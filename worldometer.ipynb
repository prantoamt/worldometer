```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
from iso3166 import countries
import json


url = "https://www.worldometers.info/coronavirus/"
req = rq.get(url)
soup = bs(req.content, "lxml")

table = soup.find("table", {"id": "main_table_countries_today"}).tbody
rows = table.find_all("tr")
data = []
Json = []

country_code_disc = {"Iran": "IR",
                    "S. Korea": "KR",
                    "UK": "GB",
                    "UAE": "AE",
                    "Russia": "RU",
                    "Vietnam": "VN",
                    "Brunei": "BN",
                    "Moldova": "MD",
                    "Bolivia": "BO",
                    "Venezuela": "VE",
                    "Ivory Coast": "CI",
                    "Saint Martin": "MF",
                    "Vatican City": "VA",
                    "St. Barth": "BL",
                    "St. Vincent Grenadines": "VC",
                    "U.S. Virgin Islands": "VI",
                    "DRC": "CD",
                    "Faeroe Islands": "FO",
                    "CAR": "CF"}

features = ["Country", "Total cases", "New cases", "Total deaths", "New deaths", 
                                   "Total recovered", "Active cases", "Serious, critical", "Tot cases/1M pop", 
                                   "Death/1M pop", "Total Tests", "Test/1M pop"]


for i in range(0,len(rows)):
    columns = rows[i].find_all("td")
    instances = []
    jdata = {}
    try:
        country_code = countries.get(columns[0].text.strip())
        country_code = country_code.alpha2
    except:
        try:
            country_code = country_code_disc[columns[0].text.strip()]
        except:
            country_code = np.nan
        
        
    instances.append(country_code)
    jdata["id"] = country_code
    for column in range(0,len(columns)):
        text = columns[column].text.strip()
        if(text==None or text ==""):
            text = np.nan
        instances.append(text)    
        jdata[features[column]] = text
    data.append(instances)
#     jdata["fill"] = "am4core.color('#A93226')"
    Json.append(jdata)
    
df = pd.DataFrame(data, columns = ["id", "Country", "Total cases", "New cases", "Total deaths", "New deaths", 
                                   "Total recovered", "Active cases", "Serious, critical", "Tot cases/1M pop", 
                                   "Death/1M pop", "Total Tests", "Test/1M pop"])
```

```python
df.to_csv(r"updated.csv", index=False)
```
